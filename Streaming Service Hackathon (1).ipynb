```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("crest")
```

```python
df = pd.read_csv('netflix_titles_nov_2019.csv', index_col='title')
```

```python
df_clean = df.copy()
```

```python
# split the Name column into three columns
df_clean[['genre_1', 'genre_2', 'genre_3']] = df['listed_in'].str.split(',',  expand=True)
```

```python
df_clean.head()
```

```python
df['country'].str.count(',').unique()
#upon checking the country data, we realise there can be up to 11 countries, 
# we will keep the first countries as being the main location for filming
df_clean['main_filming_location'] = df_clean['country'].str.split(',').str[0]
```

```python
df_clean['cast'].str.count(',').unique()
#we decide to keep the first 3 actors under cast as there can be up to 49 per row, 
#we believe the first actor would be the most famous ones
df_clean['cast_1'] = df_clean['cast'].str.split(',').str[0]
df_clean['cast_2'] = df_clean['cast'].str.split(',').str[1]
df_clean['cast_3'] = df_clean['cast'].str.split(',').str[2]
df_clean.head()
```

```python
df_clean['director'].str.count(',').unique()
#we decide to keep the first 2 directors
df_clean['director_1'] = df_clean['director'].str.split(',').str[0]
df_clean['director_2'] = df_clean['director'].str.split(',').str[1]
df_clean.head()
```

```python
#we want to split the duration column between two columns, seasons and minutes 
df_clean['duration'] = df_clean['duration'].str.split(' ').str[0]

duration_season = []
for row, value in zip(df_clean['type'], df_clean['duration']):
    if row == 'TV Show':
        duration_season.append(value)
    else: 
        duration_season.append('Nan')
df_clean['duration_season'] = duration_season


duration_minutes = []
for row, value in zip(df_clean['type'], df_clean['duration']):
    if row == 'Movie':
        duration_minutes.append(value)
    else: 
        duration_minutes.append('Nan')

df_clean['duration_minutes'] = duration_minutes
```

```python
df_clean = df_clean.drop(['show_id', 'description', 'listed_in', 'country', 'cast', 'director'], axis=1)
```

1. Split films and series, by country (country of origin)

```python
df_clean['type'].value_counts().sort_values().plot(kind='bar') 
```

2. How many series do shows run for

```python
TV_shows = df_clean[df_clean['type'] == 'TV Show']
TV_shows
TV_shows['duration_season'].value_counts().plot(kind='bar') 
```

4. Split of ratings

```python
plt.figure(figsize=(8, 4))
sns.countplot(data = df_clean, x = 'rating', order=df_clean['rating'].value_counts().index)

```

6. Average duration of films

```python
movies = df_clean[df_clean['type'] == 'Movie']
movies['duration_minutes'] = movies['duration_minutes'].astype(int)

average_movie_time = round(movies['duration_minutes'].mean())

print(f'the average time for a movie is {average_movie_time} minutes')
```

7. Top genres - adventure, action, drama, comedy, thriller

```python
genre = pd.concat([df_clean['genre_1'], df_clean['genre_2'], df_clean['genre_3']], ignore_index=True)

genre = genre.dropna()
genre = genre.apply(lambda x: x.strip())

genre.value_counts().head(10)
```

8. Which countries produce the most content

```python
df_clean['main_filming_location'].value_counts().head(10)

print(TV_shows['main_filming_location'].value_counts().head(10))
print(movies['main_filming_location'].value_counts().head(10))
```

9. Top directors globally

```python
directors = pd.concat([df_clean['director_1'], df_clean['director_2']], ignore_index=True)

directors = directors.dropna()
directors = directors.apply(lambda x: x.strip())

directors.value_counts().head(10)
```

```python
#  for TV Shows

directors_shows = pd.concat([TV_shows['director_1'], TV_shows['director_2']], ignore_index=True)

directors_shows = directors_shows.dropna()
directors_shows = directors_shows.apply(lambda x: x.strip())

directors_shows.value_counts().head(10)
```

```python
directors_movies = pd.concat([movies['director_1'], movies['director_2']], ignore_index=True)

directors_movies = directors_movies.dropna()
directors_movies = directors_movies.apply(lambda x: x.strip())

directors_movies.value_counts().head(10)
```

10. Top actors

```python
cast = pd.concat([df_clean['cast_1'], df_clean['cast_2'], df_clean['cast_3']], ignore_index=True)

cast = cast.dropna()
cast = cast.apply(lambda x: x.strip())
# I have done top 12 as there are several actors with a count of 12 appearances
cast.value_counts().head(12)
```

```python
# genre most_produced per most popular country
most_popular_genre_movies = movies_main_location['genre_1'].value_counts().head(5).index.tolist()
most_popular_genre_movies
popular_country_genre_movies = movies_main_location[movies_main_location["genre_1"].isin(most_popular_genre_movies)]
ax = sns.countplot(data = popular_country_genre_movies, x = 'genre_1', order=popular_country_genre_movies['genre_1'].value_counts().index)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.title('Most Popular Movie Genre in our Top 5 Countries')
plt.xlabel('Genre')
plt.show()
```
